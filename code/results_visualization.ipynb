```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
RMSE_results = pd.read_csv(r'.\metrics_results\RMSE_result.csv')
MAPE_results = pd.read_csv(r'.\metrics_results\MAPE_result.csv')
Jensen_Shannon_Divergence_results = pd.read_csv(r'.\metrics_results\Jensen_Shannon_Divergence_result.csv')
Kullback_Leibler_Divergence_results = pd.read_csv(r'.\metrics_results\Kullback_Leibler_Divergence_result.csv')
```

```python
plt.figure(figsize=(10, 5))
bars = plt.bar(RMSE_results['percent'], RMSE_results['Value'], color='black')

plt.xlabel('Sample_Fraction')
plt.ylabel('Results')
plt.title('RMSE_results')
plt.xticks(rotation=45) 

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, format(yval, '.4f'), va='bottom', ha='center')

plt.show()
```

```python
plt.figure(figsize=(7, 4))
bars = plt.bar(RMSE_results['percent'], RMSE_results['comparing_time_in_second'], color='black', width=0.6)

plt.xlabel('Sample_Fraction')
plt.ylabel('Time(S)')
plt.title('RMSE_Computing_Time')
plt.xticks(rotation=45)  

step = 300 / 6  
plt.yticks(np.arange(0, 400 + step, step))

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, format(yval, '.0f'), va='bottom', ha='center') 

plt.show()
```
